```python
import os
import requests
import csv
import argparse
import re
from datetime import datetime
from datetime import timedelta
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# ahir
yesterday = datetime.now() - timedelta(days=1)
fecha = yesterday.strftime('%Y-%m-%d')

```

```python
# FormulaTV
# Audiencies 27-03-2020

url = 'https://www.formulatv.com/audiencias/'+fecha

req = requests.get(url)
soup = BeautifulSoup(req.text, "lxml")

# Busquem els ul
ul = soup.h1.find_next_sibling("ul")
```

```python

# Busquem el nom de les cadenes
# Fem servir regex i esborrar la parala logo
cadenas1 = (" ".join([img['alt'] for img in ul.find_all('img', alt=True)]))

cadenas1 = re.sub(re.compile('logo'),"",cadenas1)

# Ho passem a llista
cadenas1 = cadenas1.split(" ")
print(cadenas1)
```

```python
# Treiem el percentatges, Ã©s a dir, el text de cada li
# Escollim posar un espai com a separador

audiencias1 = ul.get_text(' ', strip=True)
# Ho passem a llista
audiencias1 = audiencias1.split(" ")
print(audiencias1)
```

```python


# audiencias  = pd.DataFrame([x.split(' ') for x in audiencias1.split('\n')], index = cadenas1

audiencias = pd.DataFrame(audiencias1, columns=['Audiencia'])
cadenas = pd.DataFrame(cadenas1, columns=['Cadena'])

df = pd.concat([cadenas, audiencias], axis=1)


df["Fecha"] = fecha[0]
df
```
