```python
import os
import requests
import csv
import argparse
import re
from datetime import datetime
from datetime import timedelta
from bs4 import BeautifulSoup
```

```python
# FormulaTV
# Audiencies 27-03-2020

url = 'https://www.formulatv.com/audiencias/2020-03-27/'
fecha = [url.split('/')[-2]]

req = requests.get(url)
soup = BeautifulSoup(req.text, "lxml")
```

```python
# Comprobem el nom del header princial
soup.h1
```

```python
# De fet, l'objecte soup és el codi de html que obtindriem amb el clic botó dret
soup
```

```python
ul = soup.h1.find_next_sibling("ul")
for li in ul.find_all("li"):
    print(li)
```

```python
ul = soup.h1.find_next_sibling("ul")

# Amb un blucle
for img in ul.find_all("img"):
     print(img['alt'])
```

```python
# Més phytonic i més bonic
cadenas1 = [img['alt'] for img in ul.find_all('img', alt=True)]

# Però volem fer servir regex i esborrar la parala logo
cadenas1 = (" ".join([img['alt'] for img in ul.find_all('img', alt=True)]))

cadenas1 = re.sub(re.compile('logo'),"",cadenas1)

# Ho passem a llista
cadenas1 = cadenas1.split(" ")
print(cadenas1)
```

```python
# Treiem el percentatges, és a dir, el text de cada li
# Escollim posar un espai com a separador

audiencias1 = ul.get_text(' ', strip=True)
# Ho passem a llista
audiencias1 = audiencias1.split(" ")
print(audiencias1)
```

```python
import pandas as pd

# audiencias  = pd.DataFrame([x.split(' ') for x in audiencias1.split('\n')], index = cadenas1

audiencias = pd.DataFrame(audiencias1, columns=['Audiencia'])
cadenas = pd.DataFrame(cadenas1, columns=['Cadena'])

df = pd.concat([cadenas, audiencias], axis=1)


df["Fecha"] = fecha[0]
df
```
